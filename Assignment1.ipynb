```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, output_path="wordcloud.png"):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.savefig(output_path)
    plt.show()
```

```python
pdf_paths = ["/Users/christinadeperro/Desktop/OIM7502/OIM7502_SP25/data/foxnews.com-Embattled Harvard University President Claudine Gay hit with more plagiarism charges.pdf", "/Users/christinadeperro/Desktop/OIM7502/OIM7502_SP25/data/reuters.com-Harvard president will submit dissertation edits after plagiarism allegations.pdf"]
for pdf_path in pdf_paths:
    text = extract_text_from_pdf(pdf_path)
    generate_wordcloud(text, output_path=f"{pdf_path.split('.')[0]}_wordcloud.png")
```

Word clouds displays the words used mostly in the articles. Article 1 (Fox News report) focuses more on Gay's plagiarism and the institution, whereas Article 2 (Reuters News report) emphasizes less on the plagiarism aspect.
